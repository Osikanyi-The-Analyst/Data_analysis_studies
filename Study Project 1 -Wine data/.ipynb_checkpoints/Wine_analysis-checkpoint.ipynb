## Appending two DataFrame together
Following lines of code are to add the two addsets to enable analysis. We first of all must _set a column to identify the kype of wine for table before appending._ 

The following lines of code aid us to achieve that purpose.

```python
#creating list with numpy to add to each dataframe 
red_color=np.repeat("red", df_red.shape[0])
white_color=np.repeat("white", df_white.shape[0])

#adding the created list to the separate DataFrame
df_red["color"]=red_color
df_white["color"]=white_color

#merging the two dataframes
wine_df=df_red.append(df_white)
wine_df.head(10)
```

```python
wine_df.info()
```

```python
wine_df.to_csv("winequality_edited.csv", index=False)
```
